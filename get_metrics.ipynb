```python
path='/content/drive/MyDrive/ISRO/'
```

```python
from image_similarity_measures.evaluate import evaluation
import numpy as np
fsim=[]
issm=[]
ssim=[]
psnr=[]
k=0
for x in os.listdir("/content/drive/MyDrive/ISRO/val_pred_ohrc"):
  try:
    # print(x)
    eval_=evaluation(org_img_path=path+"high_res/"+x, pred_img_path=path+'val_pred_ohrc/'+x, metrics=["fsim", "psnr","ssim"])

    
  except:
    continue
  if eval_ is not None:
    fsim.append(eval_['fsim'])
    issm.append(eval_['issm'])
    ssim.append(eval_['ssim'])
    psnr.append(eval_['psnr'])
  
  k+=1
  print(k)
print("Average PSNR is {}".format(np.average(psnr)))
print("Average ISSM is {}".format(np.average(issm)))
print("Average FSIM is {}".format(np.average(fsim)))
print("Average SSIM is {}".format(np.average(ssim)))
```
