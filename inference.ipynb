### Slicing tif files into small images 
Each tif file from folder is sliced into small png images and then saved to new folder 

```python
path = 'add_the_path_to_the_folder_containing_test_images(tiff_file)'
inference_path = 'path_where_to_save_inferenced_files'
size = 256
```

```python
filenames = [filename for filename in glob.glob(path + '/*.tif')]
for filename in filenames:
  tmc_path = filename
  tmc_img = rasterio.open(tmc_path)
  band = tmc_img.read(1)
  band = band/(band.max()-band.min()) - band.min()/(band.max()-band.min())
  band = 255*band

  name, ext = os.path.splitext(filename.split('/')[-1])
  os.mkdir(path + '/' + name)
  os.mkdir(inference_path + '/' + name  '_inference')

  N=band.shape[1]//size
  M=band.shape[0]//size
  for i in range(0,N):
    for j in range(0,M):
      # plt.imshow(image[i:i+256,j:j+256], cmap = "gray")
      # plt.savefig('images/{}_{}.png'.format(folder[15:],N*i+j))
      im = Image.fromarray(band[i*size:i*size+size,j*size:j*size+size],mode='L')
      im.save(path + '/' + name + '/{}_{}.png'.format(i,j))
```

### Taking out inference
Each png image from the folder is taken and is passed into the model and inferenced image is stored into the `inference_folder` (given by inference_path)

Also upload final3.sh in the current folder from the drive folder submitted

```python
weights_path = 'path_to_the_weights_of_model'
```

```python
ls = os.listdir(path)
dir_ls = []
for dir in ls:
  if(os.isdir(path + '/' + dir)):
    dir_ls.append(dir)

for dir in dir_ls:
  input_read = path + '/' + dir
  output_path = inference_path + '/' + dir + '/'
  os.system('final3.sh {} {} {}'.format(input_read, weights_path, output_path))
```
